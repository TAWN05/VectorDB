```python
from pypdf import PdfReader
import chromadb
from langchain_core.documents import Document
from langchain_chroma import Chroma
import ollama 
from langchain.text_splitter import CharacterTextSplitter
from sentence_transformers import SentenceTransformer
from langchain_huggingface import HuggingFaceEmbeddings
from chromadb.utils import embedding_functions
from langchain.document_loaders import PyPDFLoader
```

```python
def strip(filename):
    # Defines text, p, joined these are all temp vars for the text stripping prosses
    text = ""
    p = ""
    joined = ""

    # Defines chunked this is my final var for the data chunks later on
    chunked = []

    # Reader is used for Pdfreader the modual im using to take the image data and turn it into plain text
    reader = PdfReader(filename)

    # This is a formality for pydpdf readed to read the pages
    for page in reader.pages:
        text += page.extract_text() + "\n"
        
    # this splits all text from "text" by "\xa0"
    readtext = text.split("\xa0")

    # Rejoins the text with spaces 
    x = " ".join(readtext)

    # Pypdf also adds "\xad" chars so they have to be striped or replaced with a nothing
    f = x.replace("\xad", "")

    """python also adds a backslash to indecate that the heiphen in "didn't" isn't the end of the string so you can strip out the heiphen 
    to not mess with embeddings (NOTE: some of the heiphens stay they dont seem to have a effect on data. These \'s can be stripped 
    out after a query)
"""    
    c = f.strip("\'")

    # chunked text is the correct chunks that is passed into the embedding function
    chunked_text = f.split('.\n')

    # this for look is the final step for cleaning up the data chunks pypdf incudes the "\n" char when at the end of a page.
    # This is a problem becuase we are sorting out chunks of data using f.split(".\n") if we remove all \n chars then the data cant be properly chunked
    for x in chunked_text:
        
        # This is used to add our final product 
        chunked += [p]

        # Defines the temp var p
        p = ""

        # This for loop maintains our chunk sizes and allows for the "\n" chars to be removed
        # Example ["This is for testing","this is the end of the page but not the end of a paragraph\n","this is the next line after"]
        # Even though we have the correct chunk sizes "\n" is still in some parts of the data
        for y in x:

            # Replaces all the left over "\n" chars to " " or a space
            p += y.replace("\n", " ")
            
            
    return chunked
```

```python

# this defines a global var from the strip(filename) fuction
chunks = strip("1540157347-HarlanEllison-IHaveNoMouthandIMustScream.pdf")
```

```python

# This creates the client we will use to do 
persistent_client = chromadb.PersistentClient()


collection = persistent_client.get_or_create_collection(name="i_have_no_mouth")
```

```python

metadatas = []
ids = []
documents = []
p = 0
n = 0
for x in chunks:
    p += 1
    f = (f"{p}")
    metadatas.append({"Book": "IAM"})
    ids.append(f"id{f}")
print("ids created: id"  f"   {f} entrys")
print("metadatas created: {Book: IAM}", f"   {p} entrys", "\n")
```

```python
# 1. Specify preffered dimensions
dimensions = 384

model = SentenceTransformer("mixedbread-ai/mxbai-embed-large-v1", truncate_dim=dimensions,)

docs = chunks
query = ""

# 2. Encode
final_embeddings = model.encode(docs)
```

```python
g = 0
for x in final_embeddings:
    g += 1
print("Created embeddings:",  g, "entrys")
#print(g)
```

```python




# ONLY RUN THIS CODE ONCE (if running this cell is required run persistaint_client.reset() this will reset the collection and clients) 



print("adding documents, embeddings, metadatas, and ids to collection...")
collection.add(
    documents = chunks,
    embeddings = final_embeddings,
    metadatas = metadatas,
    ids = ids,
)
print("data added.")
```
